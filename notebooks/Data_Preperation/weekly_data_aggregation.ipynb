```python
print(f"Before removing duplicates: {len(df):,} rows")
# Remove duplicates
df = df.drop_duplicates()
print(f"After removing duplicates: {len(df):,} rows")

# Convert date
df['week_date'] = pd.to_datetime(df['WeekID'])
df['year'] = df['week_date'].dt.year
df['month'] = df['week_date'].dt.month

def get_top_3_genres(series):
    # Count how many times each genre appears
    genre_counts = series.value_counts()

    # Pick the top 3 most common genres
    top_genres = genre_counts.index[:3]

    # Turn the list into one string like "Pop, Rock, Jazz"
    top_genres_string = ", ".join(top_genres)

    return top_genres_string

weekly_data = df.groupby('week_date').agg({
    'danceability': 'mean',
    'energy': 'mean',
    'valence': 'mean',
    'acousticness': 'mean',
    'tempo': 'mean',
    'loudness': 'mean',
    'speechiness': 'mean',
    'instrumentalness': 'mean',
    'liveness': 'mean',
    'key': 'mean',
    'mode': 'mean',
    
    # Lyrics sentiment (if you have these columns)
    'neg': 'mean',
    'neu': 'mean',
    'pos': 'mean',
    'compound': 'mean',
    
    'Song': 'count',  # Number of songs
    'spotify_track_explicit': 'mean',  # remember to drop later
    'Peak Position': 'mean',
    'Weeks on Chart': 'mean',
    
    # Most common genre
    'tag': get_top_3_genres,
    
    # Keep year/month
    'year': 'first',
    'month': 'first',
    
}).reset_index()

# Rename
weekly_data = weekly_data.rename(columns={
    'Song': 'num_songs_on_chart'
})

print(f"\n=== WEEKLY DATASET CREATED ===")
print(f"Total weeks: {len(weekly_data):,}")
print(f"Date range: {weekly_data['week_date'].min()} to {weekly_data['week_date'].max()}")

print("\n=== SAMPLE WEEKLY DATA ===")
print(weekly_data[['week_date', 'danceability', 'energy', 'valence', 
                   'pos', 'compound', 'num_songs_on_chart']].head(10))

# Save
weekly_data.to_csv('weekly_aggregated.csv', index=False)
print(f"finished: {len(df):,} rows")
print("\n✓ Saved: weekly_aggregated.csv")
```

Assign temporal features lilke seasons. Holiday Seasons matching.

```python
weekly_data['day_of_year'] = weekly_data['week_date'].dt.dayofyear
weekly_data['week_of_year'] = weekly_data['week_date'].dt.isocalendar().week
weekly_data['quarter'] = weekly_data['week_date'].dt.quarter

# Season
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

weekly_data['season'] = weekly_data['month'].apply(get_season)

# Special periods
weekly_data['is_holiday_season'] = weekly_data['month'].isin([11, 12]).astype(int)


print("\n✓ Added temporal features")
weekly_data.to_csv('weekly_aggregated.csv', index=False)
```

```python
# ========================================
# ADD EVENT LABELS
# ========================================

print("\n=== ADDING EVENT LABELS ===")

# Recession periods (NBER official dates)
recessions = [
    ('1969-12-01', '1970-11-01', 'recession_1970'),
    ('1973-11-01', '1975-03-01', 'recession_1973'),
    ('1980-01-01', '1980-07-01', 'recession_1980'),
    ('1981-07-01', '1982-11-01', 'recession_1982'),
    ('1990-07-01', '1991-03-01', 'recession_1990'),
    ('2001-03-01', '2001-11-01', 'recession_2001'),
    ('2007-12-01', '2009-06-01', 'recession_2008'),
    ('2020-02-01', '2020-04-01', 'recession_covid'),
]

# Initialize columns
weekly_data['is_recession'] = 0
weekly_data['event_name'] = None

# Mark recessions
for start, end, name in recessions:
    mask = (weekly_data['week_date'] >= start) & (weekly_data['week_date'] <= end)
    weekly_data.loc[mask, 'is_recession'] = 1
    weekly_data.loc[mask, 'event_name'] = name

# SCRATCHED, ONLY RECESSION/ECONOMIC FOCUS
violent_events = {
    '2001-09-11': '9_11_attacks',
    '2012-12-14': 'sandy_hook',
    '2013-04-15': 'boston_marathon',
    '2017-10-01': 'las_vegas_shooting',
    '2020-05-25': 'george_floyd',
}

weekly_data['is_violent_event'] = 0

for event_date, event_name in violent_events.items():
    event_dt = pd.to_datetime(event_date)
    # Find closest week
    week_diff = (weekly_data['week_date'] - event_dt).abs()
    closest_idx = week_diff.idxmin()
    weekly_data.loc[closest_idx, 'is_violent_event'] = 1
    if pd.isna(weekly_data.loc[closest_idx, 'event_name']):
        weekly_data.loc[closest_idx, 'event_name'] = event_name

# Natural disasters
natural_disasters = {
    '2005-08-29': 'hurricane_katrina',
    '2012-10-29': 'hurricane_sandy',
    '2017-08-25': 'hurricane_harvey',
    '2020-03-11': 'covid_pandemic_declared',  # WHO declaration
}

weekly_data['is_natural_disaster'] = 0

for event_date, event_name in natural_disasters.items():
    event_dt = pd.to_datetime(event_date)
    week_diff = (weekly_data['week_date'] - event_dt).abs()
    closest_idx = week_diff.idxmin()
    weekly_data.loc[closest_idx, 'is_natural_disaster'] = 1
    if pd.isna(weekly_data.loc[closest_idx, 'event_name']):
        weekly_data.loc[closest_idx, 'event_name'] = event_name

# Create overall event type
def categorize_event(row):
    if row['is_recession']:
        return 'economic_crisis'
    elif row['is_violent_event']:
        return 'violent_event'
    elif row['is_natural_disaster']:
        return 'natural_disaster'
    else:
        return 'normal'

weekly_data['event_type'] = weekly_data.apply(categorize_event, axis=1)

# ========================================
# SUMMARY
# ========================================

print("\n=== EVENT SUMMARY ===")
print(f"Total weeks: {len(weekly_data)}")
print(f"Recession weeks: {weekly_data['is_recession'].sum()}")
print(f"Violent event weeks: {weekly_data['is_violent_event'].sum()}")
print(f"Natural disaster weeks: {weekly_data['is_natural_disaster'].sum()}")
print(f"Normal weeks: {(weekly_data['event_type'] == 'normal').sum()}")

print("\n=== EVENT TYPE DISTRIBUTION ===")
print(weekly_data['event_type'].value_counts())

print("\n=== SAMPLE EVENTS ===")
events_only = weekly_data[weekly_data['event_type'] != 'normal'][['week_date', 'event_name', 'event_type', 'valence', 'energy']].head(15)
print(events_only)

# Save
weekly_data.to_csv('weekly_with_events.csv', index=False)
print("\n✓ Saved: weekly_with_events.csv")
```
