```python
import pandas as pd
import numpy as np

charts = pd.read_csv(r"C:\Users\joshu\OneDrive\Desktop\CS74\Final_Project\Hot Stuff.csv")

print(f"\nTotal rows: {len(charts):,}")
print(f"Columns: {charts.columns.tolist()}")
print(f"\nMissing values:\n{charts.isnull().sum()}")

# dates!!!!
print(f"\nDate range: {charts['WeekID'].min()} to {charts['WeekID'].max()}")
print(f"Number of unique weeks: {charts['WeekID'].nunique()}")

# how many unique songs
print(f"\nUnique songs (SongID): {charts['SongID'].nunique():,}")
print(f"Unique performers: {charts['Performer'].nunique():,}")


# 6. Sample inspection
print("\n=== First few rows ===")
print(charts.head())

print("\n=== Random song's full chart run ===")
sample_song = charts['SongID'].value_counts().index[10]  # Pick 10th most common
print(charts[charts['SongID'] == sample_song].sort_values('WeekID'))
```

```python
#audio features
audio = pd.read_csv(r"C:\Users\joshu\OneDrive\Desktop\CS74\Final_Project\Hot 100 Audio Features.csv")

# 1. Shape and completeness
print(f"\nTotal rows: {len(audio):,}")
print(f"Columns: {audio.columns.tolist()}")
print(f"\nMissing values:\n{audio.isnull().sum()}")

# 2. Unique songs
print(f"\nUnique Performer-Song combinations: {len(audio):,}")

# 3. Feature distributions
print("\n=== Audio Feature Statistics ===")
feature_cols = ['danceability', 'energy', 'valence', 'tempo', 
                'acousticness', 'loudness', 'speechiness', 
                'instrumentalness', 'liveness']
print(audio[feature_cols].describe())

# 4. Categorical features
print(f"\nExplicit tracks: {audio['spotify_track_explicit'].value_counts()}")
print(f"\nMode distribution: {audio['mode'].value_counts()}")
print(f"\nTime signatures: {audio['time_signature'].value_counts()}")

# 5. Genre inspection (critical - may need cleaning)
print(f"\nGenre field sample:")
print(audio['spotify_genre'].value_counts().head(20))
print(f"\nSongs with multiple genres: {audio['spotify_genre'].str.contains(',', na=False).sum()}")

# 6. Check for duplicates
print(f"\nDuplicate Performer-Song pairs: {audio.duplicated(subset=['Performer', 'Song']).sum()}")
```
