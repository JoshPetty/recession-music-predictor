This Notebook will be used to create basic visuals such as sentiment of music overtime.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates

```

```python
weekly_data = pd.read_csv(r'C:\Users\joshu\OneDrive\Desktop\CS74\Final_Project\weekly_aggregated.csv')
print(weekly_data.shape)

print(weekly_data.head(5))
```

```python

# Make sure it's datetime
weekly_data["week_date"] = pd.to_datetime(weekly_data["week_date"])

# Plot weekly data
fig, ax = plt.subplots(figsize=(15, 7))
ax.plot(weekly_data["week_date"], weekly_data["neg"], label="neg", color   ='red')
ax.plot(weekly_data["week_date"], weekly_data["pos"], label="pos", color   ='green')


# Format x-axis to show years only
ax.xaxis.set_major_locator(mdates.YearLocator(10))        # Major ticks every year
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y')) # Show only the year label



plt.legend()
plt.title("Weekly Sentiment Over Time (Yearly X-Axis)")
plt.xlabel("Year")
plt.ylabel("Sentiment Score")
plt.grid(True)
plt.tight_layout()
plt.show()
```
